```python
X_train = X[0:500]
y_train = y[0:500]
X_test = X[500:]
y_test = y[500:]
```

```python
mean = X_train.mean(axis = 0)
std = X_train.std(axis = 0)
```

```python
#min = X_train.min(axis = 0)
#max = X_train.max(axis = 0)
```

```python
std==0
```

```python
def remove_zero(x):
    if x == 0: return 1
    else: return x
```

```python
std_new = []
for x in std:
    y = remove_zero(x)
    std_new.append(y)
```

```python
X_train = (X_train-mean)/std_new
X_test = (X_test-mean)/std_new
```

```python
import keras
from keras import models
from keras import layers
from keras import initializers
from keras import regularizers
keras.utils.set_random_seed(1337)
```

```python
from keras.losses import MeanAbsolutePercentageError
mape = MeanAbsolutePercentageError()
```

```python
#from keras.losses import Huber
#huber_loss = Huber(delta = 3)
```

```python
#from keras.losses import LogCosh

# Define Log-Cosh loss function
#log_cosh_loss = LogCosh()
```

```python
def build_model():
    import random
    model = models.Sequential()
    model.add(layers.Dense(16, activation = 'relu',kernel_regularizer = regularizers.l1_l2(l1=0.0001,l2=0.001), input_shape = (X_train.shape[1],)))
    model.add(layers.Dense(64, activation = 'relu',kernel_regularizer = regularizers.l1_l2(l1=0.0001,l2=0.001)))
    model.add(layers.Dense(16, activation = 'relu',kernel_regularizer = regularizers.l1_l2(l1=0.0001,l2=0.001)))
    model.add(layers.Dense(1,activation = 'relu',kernel_regularizer = regularizers.l1_l2(l1=0.0001,l2=0.001))) # because output is strict;y positive
    model.compile(optimizer = 'adam',loss = mape, metrics = 'mae')
    return model
```

```python
model1 = build_model()
```

```python
model1.fit(X_train,y_train,epochs =2000,batch_size =250 )
```

```python
y_pred = model1.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.plot( range(len(y_test)),y_test,color = 'red')
plt.plot( range(len(y_test)),y_pred,color = 'blue')
```

```python
model1.evaluate(X_test,y_test)
```
